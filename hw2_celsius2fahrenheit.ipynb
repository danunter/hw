```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# Set random seed for reproducibility
np.random.seed(42)

# Create a vector of 10 random temperatures in Celsius
celsius_temps = np.round(np.random.uniform(low=-10.0, high=30.0, size=10))

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Convert Celsius temperatures to Fahrenheit
fahrenheit_temps = np.vectorize(celsius_to_fahrenheit)(celsius_temps)

# Display the Celsius and Fahrenheit temperatures
print("Celsius Temperatures:", celsius_temps)
print("Fahrenheit Temperatures:", fahrenheit_temps)
```

```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense

# Generate a 10-value vector of random temperatures in Celsius with one digit after the decimal point
celsius_temperatures = np.round(np.random.uniform(low=-10.0, high=30.0, size=10), 1)

# Convert Celsius to Fahrenheit
fahrenheit_temperatures = (celsius_temperatures * 9/5) + 32

# Create a simple neural network with one layer and one neuron
model = Sequential()
model.add(Dense(units=1, input_shape=(1,)))

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss='mean_squared_error')

# Display the model summary
model.summary()

# Reshape the data for training
celsius_temperatures = celsius_temperatures.reshape((-1, 1))
fahrenheit_temperatures = fahrenheit_temperatures.reshape((-1, 1))

# Fit the model with 500 epochs and batch size of 1
model.fit(celsius_temperatures, fahrenheit_temperatures, epochs=500, batch_size=1, verbose=1)
```

**display the weights**

```python
# Get and print the weights of the model
weights = model.get_weights()
print("Weights:")
print(weights)
```

**prediction**

```python

# Create a new Celsius vector for prediction
new_celsius = np.array([20.0, 25.0, 30.0, 35.0, 40.0])

# Reshape the data for prediction
new_celsius = new_celsius.reshape((-1, 1))

# Make predictions using the trained model
predicted_fahrenheit = model.predict(new_celsius)

# Display the predicted Fahrenheit temperatures
for c, f in zip(new_celsius.flatten(), predicted_fahrenheit.flatten()):
    print(f"{c:.1f}°C  =>  {f:.1f}°F")
```
